# plot()
일반적으로 단순히 plot을 호출할 때 2개의 1차원 배열인 X와 Y를 인수로 넣는다.  
X를 생략하면 배열 [0, 1, 2, ..., N-1]이라고 가정한다. 이 때 N은 Y의 길이다.  
즉 길이가 같은 X와 Y에서 점들의 쌍을 구하여 그래프에 표시한다. X의 각 값은 Y의 해당값과 대응하여 (x, y) 점을 얻는다.

```python
plt.plot([0, 1, 2, 3, 4], [1, 2, 4, 5, 3])
plt.show()
```

```python
A = np.linspace(0, 2 * np.pi) ## 0 ~ (2 * pi)까지 N개의 값으로 쪼갬. N의 기본값은 50이다.
print(A.shape, '\n', A)

plt.plot(A, np.sin(A)) ## A는 X축에 위치하는 값 50개, 두번째 배열은 A 각 항목의 사인 값을 지니고 있다.
plt.show()
```

```python
B = np.linspace(0.1, 10)
print(B.shape, '\n', B)

B_inverse = 1 / B
print(B_inverse)
```

- plot() 함수는 X, Y 쌍의 배열 개수 제약이 없다. 각 쌍으로부터 (x, y)의 각 항목과 같은 순서로 X, Y를 인수로 넣는다.
- fmt 인수는 그래프의 색과 스타일을 정하여 구분하는 데 유용하다.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(0, 2 * np.pi)
# plt.plot(A, np.sin(A))
# plt.plot(A, np.cos(A))
plt.plot(A, np.sin(A), A, np.cos(A))
plt.show()
```

추가적으로 fmt 인수는 그래프의 색과 스타일을 정하여 구분하는데 유용하다.

```python
plt.plot(A, np.sin(A), '^', A, np.cos(A), 'o')
plt.show()
```

- title() 함수는 그래프 전체의 제목을 지정하는데 사용된다.
- xlabel과 ylabel 함수는 각축의 구체적인 라벨을 지정하는데 사용.
- fmt에는 모든 포매팅 문자를 사용할 수 있는데 이 문자들은 "og"와 같이 조합할 수 있으며, 초록색 원을 사용하여 그래프를 그린다는 의미다.

```python
# plt.plot(A, np.sin(A), "ro", A, np.cos(A), "go")
plt.plot(A, np.sin(A), "r.", A, np.cos(A), "g>")
plt.title("Sin & Cos")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

# hist()
히스토그램은 데이터가 하위 범위에 포함되는 빈도를 보여준다. plt.hist()  
bins 인수는 하위 범위 혹은 통의 개수를 정의한다. 기본 설정은 10이며, 가장 큰 값부터 가장 작은 값의 차이를 10으로 나눠서 하위 범위의 크기를 정한다.

```python
import matplotlib.pyplot as plt

weight_list = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
max_weight, min_weight = max(weight_list), min(weight_list)
print(max_weight, min_weight)
print(max_weight - min_weight)

weight_arr = np.array(weight_list)

plt.hist(weight_arr, bins=25)
plt.title("Weights")
plt.show()
```

np.histogram() 함수는 히스토그램 결과를 담은 신규 배열을 만든다. 신규 배열의 각 항목은 각통에 대응하는 숫자의 빈도이다.

```python
A = np.random.standard_normal(200000)
A = A * 10 + 100
B, X = np.histogram(A, 50)

print(np.array(B).shape, np.array(X).shape)
print(B[:10], X[:10])

X = (X[1:] + X[:-1]) / 2 ## 경계 대신 통위 중간 값을 사용한다.
plt.plot(X, B)
plt.show()
```
